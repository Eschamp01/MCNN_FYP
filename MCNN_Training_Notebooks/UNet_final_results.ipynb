```python
import pandas as pd 

unet_dice_1 = pd.read_csv(root_dir+project_folder+"UNet_600_dice.csv")
unet_dice_2 = pd.read_csv(root_dir+project_folder+"UNet_600-800_dice.csv") 
```

```python
import matplotlib.pyplot as plt
```

```python
# UNet Backbone

val_interval = 2

plt.figure("UNet dice", (6, 6))

plt.title("Medical UNet Val Mean Dice", fontsize=18)
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("Dice metric", fontsize=14)

x1 = [val_interval * (i + 1) for i in range(len(unet_dice_1.iloc[:,1]))]
y1 = unet_dice_1.iloc[:,1]
plt.plot(x1, y1,'b')

x2 = [val_interval * (i + 301) for i in range(len(unet_dice_2.iloc[:,1]))]
y2 = unet_dice_2.iloc[:,1]
plt.plot(x2, y2,'b')

plt.legend()
plt.show()
```
