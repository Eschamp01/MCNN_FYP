# Load losses and dice metrics as data frames

```python
import pandas as pd

low_loss = pd.read_csv(root_dir+project_folder+"MCNN_low_res_0-400_loss.csv") 
med_100_loss = pd.read_csv(root_dir+project_folder+"100_med_loss.csv") 
med_200_loss = pd.read_csv(root_dir+project_folder+"200_med_loss.csv") 
med_300_loss = pd.read_csv(root_dir+project_folder+"300_med_loss.csv") 
med_400_loss = pd.read_csv(root_dir+project_folder+"400_med_loss.csv") 
high_100_loss = pd.read_csv(root_dir+project_folder+"100_high_loss_1.csv") 
high_100_loss_2 = pd.read_csv(root_dir+project_folder+"100_high_loss_2.csv") 
high_200_loss = pd.read_csv(root_dir+project_folder+"200_high_loss_1.csv") 
high_300_loss = pd.read_csv(root_dir+project_folder+"300_high_loss_1.csv") 

low_dice = pd.read_csv(root_dir+project_folder+"MCNN_low_res_0-400_dice.csv") 
med_100_dice = pd.read_csv(root_dir+project_folder+"100_med_dice.csv") 
med_200_dice = pd.read_csv(root_dir+project_folder+"200_med_dice_1.csv") # 200_med_dice_1 also exists?
med_300_dice = pd.read_csv(root_dir+project_folder+"300_med_dice.csv") 
med_400_dice = pd.read_csv(root_dir+project_folder+"400_med_dice.csv") 
high_100_dice = pd.read_csv(root_dir+project_folder+"100_high_dice_1.csv") 
high_100_dice_2 = pd.read_csv(root_dir+project_folder+"100_high_dice_2.csv") 
high_200_dice = pd.read_csv(root_dir+project_folder+"200_high_dice_1.csv") 
high_300_dice = pd.read_csv(root_dir+project_folder+"300_high_dice_1.csv") 
```

# Graph Losses and dice metrics

```python
import matplotlib.pyplot as plt
```

```python
nums = low_loss.iloc[:,0]
full_res_loss = low_loss.iloc[:,2]
```

```python
# FocusNet100

val_interval = 2

plt.figure("train", (12, 6))

plt.subplot(1, 2, 1)
plt.title("FocusNet100 Val Loss", fontsize=18)
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("Dice loss", fontsize=14)

x1 = [i + 1 for i in range(len(low_loss.iloc[:,2][:100]))]
y1 = low_loss.iloc[:,2][:100]
plt.plot(x1, y1)

x2 = [i + 101 for i in range(len(med_100_loss.iloc[:,2]))]
y2 = med_100_loss.iloc[:,2]
plt.plot(x2, y2)

x2 = [i + 201 for i in range(len(high_100_loss.iloc[:,2][:350]))]
y2 = high_100_loss.iloc[:,2][:350]
plt.plot(x2, y2, 'g')

x3 = [i + 551 for i in range(len(high_100_loss_2.iloc[:,2]))]
y3 = high_100_loss_2.iloc[:,2]
plt.plot(x3, y3, 'g')

plt.subplot(1, 2, 2)
plt.title("FocusNet100 Val Mean Dice", fontsize=18)
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("Dice metric", fontsize=14)

x1 = [val_interval * (i + 1) for i in range(len(low_dice.iloc[:,1][:50]))]
y1 = low_dice.iloc[:,1][:50]
plt.plot(x1, y1, label="Low Res Loss")

x2 = [val_interval * (i + 51) for i in range(len(med_100_dice.iloc[:,1]))]
y2 = med_100_dice.iloc[:,1][:50]
plt.plot(x2, y2, label="Medium Res Loss")

x3 = [val_interval * (i + 101) for i in range(len(high_100_dice.iloc[:,1][:176]))]
y3 = high_100_dice.iloc[:,1][:176]
plt.plot(x3, y3, 'g', label="High Res Loss")

x3 = [val_interval * (i + 276) for i in range(len(high_100_dice_2.iloc[:,1]))]
y3 = high_100_dice_2.iloc[:,1]
plt.plot(x3, y3, 'g')

plt.legend()
plt.show()
```
